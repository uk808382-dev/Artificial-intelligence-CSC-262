```python
# Simple Hill Climbing Algorithm

def hill_climbing():

    # Sample values (higher is better)
    states = [2, 5, 8, 12, 7, 3]

    current = 0   # start from first position

    while current < len(states) - 1:

        # Move to next state if it is better
        if states[current + 1] > states[current]:
            current += 1
        else:
            break

    print("Best state value:", states[current])
    print("Position:", current)

# Run program
hill_climbing()
```

Hill Climbing Algorithm – Questions & Answers
1. What is the purpose of this program?

Answer: It finds the best (highest value) state using a simple hill climbing approach.

2. What does the hill climbing algorithm do?

Answer: It improves a solution step by step by moving toward better neighboring states.

3. What is stored in the states list?

Answer: It contains different values representing possible solution states.

4. What is the starting point of the algorithm?

Answer: The algorithm starts from index 0 of the list.

5. Why is current = 0 used?

Answer: It sets the initial position at the first state in the list.

6. What is the condition in the while loop?

Answer: current < len(states) - 1 ensures the program does not go out of bounds.

7. What does states[current + 1] represent?

Answer: It represents the next neighboring state.

8. What comparison is used in the algorithm?

Answer: It compares the current state with the next state to check which is better.

9. What happens if the next state is better?

Answer: The algorithm moves to that next state.

10. What happens if the next state is not better?

Answer: The loop stops execution.

11. What is the final output of the program?

Answer: It prints the best value found and its position in the list.

12. Why is this called a greedy approach?

Answer: Because it always chooses the immediate best option without checking future states.

13. What is a limitation of this algorithm?

Answer: It may stop at a local maximum instead of the global maximum.

14. What is a local maximum?

Answer: A state that is better than its neighbors but not the best overall.

15. What is a global maximum?

Answer: The highest possible value in the entire search space.

16. Can this algorithm always find the best solution?

Answer: No, it may miss better solutions due to its greedy nature.

17. What type of search is hill climbing?

Answer: It is a heuristic search technique.

18. What is the role of the while loop?

Answer: It repeatedly checks for better states until no improvement is found.

19. What does current += 1 do?

Answer: It moves the algorithm to the next state in the list.

20. Why is this called simple hill climbing?

Answer: Because it only checks the next immediate neighbor without backtracking or exploring multiple paths.
